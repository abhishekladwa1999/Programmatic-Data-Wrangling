# Insight 1: finding a link between original language and revenue of Films

As my first insight, I wanted to look into a potential (and perhaps
somewhat overlooked and maybe even unlikely) link between the revenue of the films, and the original language that the film was produced in.

After drawing up a list of original languages and comparing them to the average revenue for the films of the that language, I could immediately some surprises.

Due to english being the most commonly spoken, and deemed 'international' language, it was expected that the English language films would top substantially in terms of revenue compared to other language. However, Telugu language films (ISO code 'te') seems to have displaced English, with a higher revenue.

This could be due to more films produced in English, and more commercial failures bringing down the overall average revenue of English language films, but it also is an interesting insight to see that Telugu films, and perhaps to a wider extent other language films, could be more financially successful, despite being less spoke or understood around the world.

Furthermore, when looking into and comparing budget, we can see that the average budget given for Telugu films, is also the highest budget given.

# Insight 2: concluding on a take away on an insight into revenue

A key finding that was found upon conducting analysis on how the revenues of all the films, was how how on average, most films in the database fit into one single range of values, specifically on the lowest end of values, compared to the revenue ranges of other films in the database.


Taking insight from the distribution plot of revenue against count, I also decided to further confirm this trend with addition of price buckets, and a 'box-office' grouping, to showcase that this had the same trend as the original distribution plot.

While an insight could initially be drawn suggesting that most of the films in the dataset produced low revenue film, upon further insight and looking into the values, we can conclude that there are multiple films that act as outliers which cause most of these films to fall into the lower band. Upon separating the distribution of revenues by film in buckets, we can also see this trend.

This concludes that most of the films, while on the lowest band of revenue, actually do portray more of stanadard average, as the majority of films fall into this revenue band, with there being several outliers in the data that fall into much larger revenue bands. An example of this would be Avatar.

```python
sns.displot(
            x = 'revenue',
            data = df,
            bins = 5
           )

plt.show()
```

```python
df[['revenue']].sort_values(by = 'revenue', ascending = False)
```

```python
box_office = df[['id','title', 'budget', 'genres', 'popularity', 'revenue']].copy()

box_office.drop_duplicates(keep = 'last', inplace = True)

revenue_buckets = [0, 500000000, 1000000000, 1500000000, 2000000000, 2500000000, 3000000000]
revenue_name = ['very low revenue', 'low revenue', 'moderate revenue', 'high revenue', 'very high revenue', 'immensely high revenue']

box_office['Price Buckets'] = pd.cut(box_office['revenue'],
                                  bins = revenue_buckets,
                                  labels = revenue_name,
                                  include_lowest = True)
```

```python
box_office.head()
```

```python
sns.countplot(x = 'Price Buckets',
              data = box_office,
              palette = 'magma')

plt.xlabel('Price Group')
plt.ylabel('Number of Films')

plt.xticks(rotation = 45,
           horizontalalignment = 'right',
           fontweight = 'light',
           fontsize = 'large',
          )
plt.title('box office')

plt.show()
```

```python
def box_office_grouping(x):
    if x <500000000:
        return '0-0.5 Bil.'
    elif x<1000000000:
        return '5-10 Bil.'
    elif x<1500000000:
        return '10-15 Bil.'
    elif x<2000000000:
        return '15-20 Bil.'
    elif x<2500000000:
        return '20-25 Bil.'
    elif x<3000000000:
        return '25-30 Bil.'
    else:
        return '30+ Bil.'

df['box_office_group'] = df['revenue'].apply(box_office_grouping)



plt.figure(figsize = (16,9))

sns.countplot(data = df,
              x = 'box_office_group',
              palette = 'viridis',
              order = ['0-0.5 Bil.','5-10 Bil.','10-15 Bil.','15-20 Bil.','20-25 Bil.','25-30 Bil.','30+ Bil.'])

plt.show()
```

# Movie 2: insights - March of the penguins

 When comparing the highest grossing film of each genre to the highest grossing film of the 'Documentary' genre: March of the Penguins, I discovered that it was the only film of the highest grossing film of its respective genre to not be in English, but rather in French.

 While at first, this may not seem very insightful, it does seems interesting that while every other film genre is dominated by the English language film market, the fact that a French documentary has outranked all other English documentaries in revenue *could* suggest less interest in documentaries and more interest towards all other genres across English languages speakers in general. However, this would definitely need to be backed up with further analysis.


```python
import ast
def extract_genres(x):
    x = ast.literal_eval(x)
    Genres = []

    for item in x:
        Genres.append(item['name'])

    return Genres

df['extracted_genres'] = df['genres'].apply(extract_genres)
```
